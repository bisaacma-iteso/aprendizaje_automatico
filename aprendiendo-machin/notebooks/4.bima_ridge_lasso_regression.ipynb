# Regularization (Ridge and Lasso)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    1, 2, 3, 4, 5, 6
]).reshape(-1, 1)

y = np.array([
    1.5, 2.8, 3.8, 3.7, 5.7, 5.5
]).reshape(-1)
```

```python
plt.scatter(X, y)
```

```python
X_train = X[0:2]
y_train = y[0:2]

X_test = X[2:]
y_test = y[2:]

plt.scatter(X_train, y_train, label="train")
plt.scatter(X_test, y_test, label="Test")
plt.legend()

print(f'{X_train.shape} {y_train.shape} {X_test.shape} {y_test.shape}')
```

```python
from sklearn.linear_model import SGDRegressor

vanilla = SGDRegressor(penalty=None)
ridge = SGDRegressor(penalty="l2", alpha=0.2)
lasso = SGDRegressor(penalty="l1", alpha=0.2)
```

```python
vanilla.fit(X_train, y_train)
vanilla_predict = vanilla.predict(X_test)
vanilla.score(X_test, y_test)
```

```python
ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
lasso.fit(X_train, y_train)
lasso.score(X_test, y_test)
```

```python
plt.scatter(X, y, c='r')
plt.plot(X, vanilla.predict(X), label="Vanilla")
plt.plot(X, ridge.predict(X), label="Ridge")
plt.plot(X, lasso.predict(X), label="Lasso")
plt.legend()
```
