# MNIST dataset (basic image classification)

```python
from sklearn.datasets import load_digits

mnist = load_digits()
mnist.keys()
```

```python
mnist.images.shape
```

```python
mnist.data.shape
```

```python
import matplotlib.pyplot as plt

index = 89
plt.imshow(mnist.images[index], cmap='gray')
plt.title(f'Number: {mnist.target[index]}')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
print(mnist.data.shape)
X_surprise = pca.fit_transform(mnist.data)
#X_surprise.shape
```

```python
plt.scatter(X_surprise[:, 0], X_surprise[:, 1], c=mnist.target, cmap='RdYlBu')
plt.colorbar()
```
